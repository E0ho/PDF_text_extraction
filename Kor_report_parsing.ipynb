```python
import cv2

# 이미지 불러오기
image = cv2.imread("./kor_sample.jpg")
image = image[..., ::-1]
```

# Image Segmentation을 위한 모델 Detectron2 적용

Detectron2.JPG

```python
import layoutparser as lp

# Model 생성
model = lp.Detectron2LayoutModel('lp://PrimaLayout/mask_rcnn_R_50_FPN_3x/config',
                                 extra_config = ["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.65],

                                 label_map = {1:"TextRegion", 2:"ImageRegion", 3:"TableRegion", 4:"MathsRegion", 5:"SeparatorRegion", 6:"OtherRegion"})
# Layout 감지
layout = model.detect(image)
```

```python
color_map = {
    'TextRegion':   'green',
    'ImageRegion':  'blue',
    'TableRegion':   'blue',
    'MathsRegion':  'blue',
    'SeparatorRegion': 'pink',
    'OtherRegion' : 'white'
}

# Image 내 Layout 시각화
lp.draw_box(image, layout, color_map=color_map)
```

```python
# 각 Label의 Layout Block 정의
text_blocks = lp.Layout([b for b in layout if b.type=="TextRegion"])
image_block = lp.Layout([b for b in layout if b.type=="ImageRegion"])
table_blocks = lp.Layout([b for b in layout if b.type=='TableRegion'])
maths_blocks = lp.Layout([b for b in layout if b.type=='MathsRegion'])
separator_blocks = lp.Layout([b for b in layout if b.type=='SeparatorRegion'])
other_blocks = lp.Layout([b for b in layout if b.type=='OtherRegion'])
```

```python
# Figure 내 Text 제거
text_blocks = lp.Layout([b for b in text_blocks \
                   if not any(b.is_in(b_fig) for b_fig in table_blocks)])
```

```python
h, w = image.shape[:2]

left_interval = lp.Interval(0, w/2*1.05, axis='x').put_on_canvas(image)

left_blocks = text_blocks.filter_by(left_interval, center=True)
left_blocks.sort(key = lambda b:b.coordinates[1])

right_blocks = [b for b in text_blocks if b not in left_blocks]
right_blocks.sort(key = lambda b:b.coordinates[1])

# And finally combine the two list and add the index
# according to the order
text_blocks = lp.Layout([b.set(id = idx) for idx, b in enumerate(left_blocks + right_blocks)])
```

```python
lp.draw_box(image,
            text_blocks,
            box_width = 5,
            color_map=color_map,
            show_element_id = True)
```

# 각 Layout 내 Text 추출을 위해 Tesseact-ocr 기술 사용

```python
!sudo apt install tesseract-ocr
```

```python
# 번역 언어 = English
ocr_agent = lp.TesseractAgent(languages='eng+kor2')
```

```python
# Box 내 Padding 정보 추가하여 글자 짤림 방지
for block in text_blocks:
    segment_image = (block.pad(left=15, right=15, top=15, bottom=15).crop_image(image))


    text = ocr_agent.detect(segment_image)
    block.set(text=text, inplace=True)
```

```python
for txt in text_blocks.get_texts():
    print(txt, end='\n---\n')
```
